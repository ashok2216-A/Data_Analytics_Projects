```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
plt.style.use('seaborn-darkgrid')
plt.rc('figure',figsize=(18,8))
plt.rc('lines',markersize=14)
plt.rc('font', size=14)
import warnings
warnings.filterwarnings('ignore')
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```
description_text = " ".join(df.description)

wordcloud = WordCloud(width=1500, height=500).generate(description_text)

plt.figure( figsize=(30,20))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title('Description')
plt.show()
```

```
listed_text = " ".join(df.listed_in)

wordcloud = WordCloud(width=1500, height=500).generate(listed_text)

plt.figure( figsize=(30,20))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title('Listed In')
plt.show()
```
